## 1. Bidirectional RNNs (Bi-RNN)
Standard RNNs only look at past context. But in many tasks (like translation or speech recognition), future context is also useful. Bi-RNNs process the sequence in both directions (Forward and Backward) and concatenate the hidden states.

## 2. Stacked (Deep) RNNs
Just like Deep Neural Networks, we can stack multiple RNN layers on top of each other. The output of the first RNN layer becomes the input to the second layer.

```python
from tensorflow.keras.layers import Bidirectional, LSTM, Dense
from tensorflow.keras.models import Sequential

# Creating a Deep Bidirectional LSTM Model
model = Sequential()
model.add(Bidirectional(LSTM(64, return_sequences=True), input_shape=(10, 50)))
model.add(Bidirectional(LSTM(32))) # Last layer doesn't need to return sequences if it connects to Dense
model.add(Dense(1, activation='sigmoid'))

model.summary()
# Note the increase in parameters due to Bidirectional (x2) and Stacking.
```

## 3. Intro to Seq2Seq
Sequence-to-Sequence models map a sequence of fixed length to another sequence of fixed length (e.g., Translation).
It consists of an **Encoder** (processes input) and a **Decoder** (generates output).
