# 05. Gated Recurrent Unit (GRU)

## Overview
GRU is a variation of the LSTM. It combines the forget and input gates into a single 'update gate'. It also merges the cell state and hidden state, and makes some other changes. The resulting model is simpler than standard LSTM models (fewer parameters) and has been becoming increasingly popular.

## Gates in GRU
1. **Update Gate ($z_t$)**: Helps the model to determine how much of the past information (from previous time steps) needs to be passed along to the future.
2. **Reset Gate ($r_t$)**: Decides how much of the past information to forget.

## LSTM vs GRU
- **GRU**: Simpler, faster to train, 2 gates.
- **LSTM**: More powerful/flexible, 3 gates, separate cell state.

There is no clear winner; it depends on the dataset.
